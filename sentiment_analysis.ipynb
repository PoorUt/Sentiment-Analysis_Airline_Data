```python
#get the number of negative reasons
data['negativereason'].nunique()

NR_Count=dict(data['negativereason'].value_counts(sort=False))
def NR_Count(Airline):
    if Airline=='All':
        a=data
    else:
        a=data[data['airline']==Airline]
    count=dict(a['negativereason'].value_counts())
    Unique_reason=list(data['negativereason'].unique())
    Unique_reason=[x for x in Unique_reason if str(x) != 'nan']
    Reason_frame=pd.DataFrame({'Reasons':Unique_reason})
    Reason_frame['count']=Reason_frame['Reasons'].apply(lambda x: count[x])
    return Reason_frame
def plot_reason(Airline):
    
    a=NR_Count(Airline)
    count=a['count']
    Index = range(1,(len(a)+1))
    plt.bar(Index,count, color=['red','yellow','blue','green','black','brown','gray','cyan','purple','orange'])
    plt.xticks(Index,a['Reasons'],rotation=90)
    plt.ylabel('Count')
    plt.xlabel('Reason')
    plt.title('Count of Reasons for '+Airline)


plot_reason('All')
plt.figure(2,figsize=(13, 13))
for i in airlines:
    indices= airlines.index(i)
    plt.subplot(2,3,indices+1)
    plt.subplots_adjust(hspace=0.9)
    plot_reason(i)
```

Now, we will clean the tweet text data and apply classification algorithms on it.

```python
def tweet_to_words(tweet):
    letters_only = re.sub("[^a-zA-Z]", " ",tweet) 
    words = letters_only.lower().split()                             
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops] 
    return( " ".join( meaningful_words ))
```

```python
nltk.download('stopwords')
data['clean_tweet']=data['text'].apply(lambda x: tweet_to_words(x))
```

```python
x = data.clean_tweet
y = data.airline_sentiment

print(len(x),len(y))
```

```python
#Train test split. 80:20

x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42)
print(len(x_train),len(y_train))
print(len(x_test), len(y_test))
```

```python
#we need the sklearn countvectorizer to convert the text data in to numericals.
from sklearn.feature_extraction.text import CountVectorizer

#Create a count vectorizer object
vect =  CountVectorizer()
vect.fit(x_train)
```

```python
#Create a document- term matrix from the train and test sets
```


A Document-Term Matrix (DTM) is a mathematical matrix that describes the frequency of terms (words) that appear in a collection of documents.   
It's commonly used in text mining and natural language processing (NLP) to analyze and represent text data.

Structure of a Document-Term Matrix
In a DTM, rows represent individual documents, and columns represent terms (words) from the entire corpus.  
 Each cell in the matrix contains the frequency (count) of a specific term in a specific document.

Example
Consider a simple corpus of three documents:

"I love programming"

"Programming in Python"

"Python is great"

The corresponding DTM would look something like this:



```python
x_train_dtm = vect.transform(x_train)
x_test_dtm = vect.transform(x_test)
```

```python
vect_tunned = CountVectorizer(stop_words='english', ngram_range=(1,2), min_df=0.1, max_df=0.7, max_features=100)
vect_tunned
```

```python
#Model Building

#training SVM model with linear kernel
#Support Vector Classification-wrapper around SVM
from sklearn.svm import SVC
model = SVC(kernel='linear', random_state = 10)
model.fit(x_train_dtm, y_train)
#predicting output for test data
pred = model.predict(x_test_dtm)
```

```python
#accuracy score
accuracy_score(y_test,pred)
```

```python
#building confusion matrix
cm = confusion_matrix(y_test,pred)
cm
```

```python
cm
```

```python
#defining the size of the canvas
plt.rcParams['figure.figsize'] = [15,8]
#confusion matrix to DataFrame
conf_matrix = pd.DataFrame(data = cm,columns = ['Predicted:0','Predicted:1',], index = ['Actual:0','Actual:1',])
#plotting the confusion matrix
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = 'Paired', cbar = True,linewidths = 0.5, annot_kws = {'size':20})
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.show()
```

```python
# print the classification report
print(classification_report(y_test,pred))
```
